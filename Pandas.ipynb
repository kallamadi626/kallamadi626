#Pandas Series

```python
info=np.array(['A','B','c','D','E','F'])
a=pd.Series(info)
print(a)
```

```python
info={'x':0,'y':1,'z':2}
a=pd.Series(info)
print(a)
```

```python
arr=[1,2,3,4]
index=['a','b','c','d']
x=pd.Series(arr,index)
print(x)
```

```python
a=pd.Series(data=[1,2,3,np.nan],index=[0,1,2,3])
print(a)
print(a.hasnans)
```

#loc and iloc

```python
data = pd.Series(['a','b','c','d'],index=[1,3,5,7])
print(data)
print(data.loc[1])
print(data.iloc[1])
```

```python
data={"Name":['A','B','C','D'],'Age':['25','21','62','41']}
df=pd.DataFrame(data)
print(df.dtypes)
df
```

```python
#astype---> convert one dtype into another dtype
df['Age']=df['Age'].astype(int)
print(df.dtypes)
df
```

```python
data={'Name':['A','B',np.nan,'D'],'Age':[25,np.nan,62,41]}
df=pd.DataFrame(data)
df
```
